    <h3 style="color: #4CAF50;"><strong style="color: #FF5733;">The border-radius Property in CSS</strong></h3>
    <p style="background-color: #D2DE32; padding: 10px; text-align: left; border: 4px solid #ECEE81; border-radius: 10px;">
        The border-radius property in CSS is used to create rounded corners for elements. You can specify different styles of rounded corners by providing different values for border-radius. Here are some ways you can use border-radius to achieve various styles:
    </p>
    <h4 style="color: #4CAF50;">Single Value</h4>
    <p style="background-color: #E0F7FA; padding: 10px; text-align: left; border: 4px solid #00ACC1; border-radius: 10px;">
        <strong>Uniform Radius:</strong> Applies the same radius to all corners.<br>
        <code>border-radius: 10px;</code>
    </p>
    <h4 style="color: #4CAF50;">Two Values</h4>
    <p style="background-color: #FFECB3; padding: 10px; text-align: left; border: 4px solid #FFC107; border-radius: 10px;">
        <strong>Horizontal and Vertical Radii:</strong> The first value applies to the top-left and bottom-right corners, and the second value applies to the top-right and bottom-left corners.<br>
        <code>border-radius: 10px 20px;</code>
    </p>
    <h4 style="color: #4CAF50;">Four Values</h4>
    <p style="background-color: #D1C4E9; padding: 10px; text-align: left; border: 4px solid #7E57C2; border-radius: 10px;">
        <strong>Individual Corner Radii:</strong> Applies different radii to each corner in the order of top-left, top-right, bottom-right, and bottom-left.<br>
        <code>border-radius: 10px 20px 30px 40px;</code>
    </p>
    <h4 style="color: #4CAF50;">Slash Notation</h4>
    <p style="background-color: #FFCDD2; padding: 10px; text-align: left; border: 4px solid #E57373; border-radius: 10px;">
        <strong>Elliptical Radius:</strong> Using the slash notation, you can specify horizontal and vertical radii separately.<br>
        <code>border-radius: 10px / 20px;</code>
    </p>
    <h4 style="color: #4CAF50;">Examples</h4>
    <p style="background-color: #DCEDC8; padding: 10px; text-align: left; border: 4px solid #8BC34A; border-radius: 10px;">
        <strong>Uniform Rounded Corners:</strong><br>
        <code>border-radius: 15px;</code><br>
        All corners will have a radius of 15px.
    </p>
    <p style="background-color: #F8BBD0; padding: 10px; text-align: left; border: 4px solid #F06292; border-radius: 10px;">
        <strong>Different Radius for Each Corner:</strong><br>
        <code>border-radius: 15px 50px 30px 5px;</code><br>
        This will apply 15px to the top-left corner, 50px to the top-right corner, 30px to the bottom-right corner, and 5px to the bottom-left corner.
    </p>
    <p style="background-color: #FFCCBC; padding: 10px; text-align: left; border: 4px solid #FF7043; border-radius: 10px;">
        <strong>Elliptical Corners:</strong><br>
        <code>border-radius: 50px / 20px;</code><br>
        This creates an elliptical shape with a horizontal radius of 50px and a vertical radius of 20px.
    </p>
    <p style="background-color: #C8E6C9; padding: 10px; text-align: left; border: 4px solid #66BB6A; border-radius: 10px;">
        <strong>Top Corners Rounded:</strong><br>
        <code>border-radius: 20px 20px 0 0;</code><br>
        Only the top-left and top-right corners will be rounded.
    </p>
    <p style="background-color: #FFECB3; padding: 10px; text-align: left; border: 4px solid #FFC107; border-radius: 10px;">
        <strong>Bottom Corners Rounded:</strong><br>
        <code>border-radius: 0 0 20px 20px;</code><br>
        Only the bottom-left and bottom-right corners will be rounded.
    </p>
    <p style="background-color: #E1BEE7; padding: 10px; text-align: left; border: 4px solid #BA68C8; border-radius: 10px;">
        <strong>Complex Elliptical Corners:</strong><br>
        <code>border-radius: 10px 20px 30px 40px / 40px 30px 20px 10px;</code><br>
        This applies separate horizontal and vertical radii to each corner, creating a more complex shape.
    </p>
